#### Data Clean up

```python
import pandas as pd
import urllib.request
import zipfile
```

```python
url = 'https://github.com/mattharrison/datasets/raw/master/data/kaggle-survey-2018.zip'
zname = 'keggle-survey-2018.zip'
fname = 'multipleChoiceResponses.csv'
 
```

```python
def extract_zip_file(src:str, file_path:str, file_name:str):
    """
    Extract the data frome zip file and read it into pandas dataframe
    parameters:
        src: URL of the zip file.
        file_path: file path where zip file will be written.
        file_name: name of the file inside the zip file
    """

    f_in = urllib.request.urlopen(src)
    data = f_in.read()


    with open(file_path, mode='wb') as f_out:
        f_out.write(data)

    with zipfile.ZipFile(file_path) as Z:
        d = pd.read_csv(Z.open(file_name))
        d_questions = d.iloc[0]
        raw = d.iloc[1:]

        return raw

```

```python
raw = extract_zip_file(url, zname, fname)
```
